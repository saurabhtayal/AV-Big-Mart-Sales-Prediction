```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'whitegrid')
pd.pandas.set_option('display.max_columns', None)
from sklearn.linear_model import Lasso, LassoCV
from sklearn.feature_selection import SelectFromModel
```

```python
X_train = pd.read_csv('X_train_scaled.csv')
display(X_train.head())
print(X_train.shape)

X_test = pd.read_csv('X_test_scaled.csv')
display(X_test.head())
print(X_test.shape)

y_df = pd.read_csv('y_train_transformed.csv', names = ['y_train_transformed'])
display(y_df.head())
print(y_df.shape)
```

```python
y_train = y_df['y_train_transformed']
display(y_train.head())
print(y_train.shape)

X_train = X_train.drop('Unnamed: 0', axis = 1)
X_test = X_test.drop('Unnamed: 0', axis = 1)

display(X_train.head())
print(X_train.shape)

display(X_test.head())
print(X_test.shape)
```

```python
np.logspace(-3, 0, 1000)
```

```python
alphas = np.logspace(-3, -2, 1000)

lassocv = LassoCV(alphas = alphas, cv = 10, max_iter = 1000, random_state = 0, \
                  precompute = True, n_jobs = -1, selection = 'random')

lassocv.fit(X_train, y_train)

print('Best Alpha Value from LassoCV:', lassocv.alpha_)
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha = lassocv.alpha_, random_state = 0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(X_train, y_train)
```

```python
print(feature_sel_model.get_support())
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_train.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format(X_train.shape[1]))
print('No. of features selected by Lasso: {}'.format(len(selected_feat)))
```

```python
print(selected_feat)
```

```python
X_train = X_train[selected_feat]
display(X_train.head())
```

```python
X_test = X_test[selected_feat]
display(X_test.head())
```

```python
X_train.to_csv('X_train_lasso.csv', index = False)
X_test.to_csv('X_test_lasso.csv', index = False)
```
